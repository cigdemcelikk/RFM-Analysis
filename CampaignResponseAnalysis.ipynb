```python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the box plots for all columns
df.boxplot(ax=ax)

# Show the plot
plt.show()
```

```python
# Gelir değeri üzerinden gruplama
threshold_income = 40000  # Gelir eşiği
# A grubu (geliri düşük olanlar)
group_A = df[df['Income'] <= threshold_income]

# B grubu (geliri yüksek olanlar)
group_B = df[df['Income'] > threshold_income]
```

```python
from scipy.stats import shapiro

# A Grubu için Shapiro-Wilk testi (yanıt değişkeni)
test_stat_A, p_value_A = shapiro(group_A['Response'])

# B Grubu için Shapiro-Wilk testi (yanıt değişkeni)
test_stat_B, p_value_B = shapiro(group_B['Response'])

# Sonuçları yazdıralım
print(f"A Grubu - Test İstatistiği: {test_stat_A:.4f}, P-değeri: {p_value_A:.4f}")
print(f"B Grubu - Test İstatistiği: {test_stat_B:.4f}, P-değeri: {p_value_B:.4f}")
```

Shapiro-Wilk testinin p-değeri 0.05'ten küçük olduğunda, verinin normal dağılımdan sapma gösterdiği kabul edilir. P-değerinin 0 olması, genellikle çok belirgin bir sapma olduğunu ve verinin normal dağılmadığını güçlü bir şekilde gösterir.Bu durumda, non-parametrik testler tercih edilmelidir.

```python
from scipy import stats

# Normal dağılmayan veriler için Mann-Whitney U Testi
u_stat, p_value_u = stats.mannwhitneyu(group_A['Response'], group_B['Response'])

print(f"U-istatistiği: {u_stat:.4f}, P-değeri: {p_value_u:.4f}")
```
